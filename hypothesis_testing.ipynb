Use the following contingency table to help answer the question of whether using a Macbook and being a Codeup student are independent of each other.

```python
data = {
    "uses_a_macbook": [49, 20],
    "doesnt_use_a_macbook": [1, 30]
}
# Setting the index names
index_names = ["codeup_student", "not_codeup_student"]
# Creating the DataFrame
observed = pd.DataFrame(data, index=index_names)
observed
```

```python
#set our alpha
alpha = 0.05
```

```python
#H0= There is no relationship between macbook users and code up students.
#Ha= There is a relationship between macbook users and code up students.
```

```python
stats.chi2_contingency(observed)
```

```python
chi2, p, dof, expected = stats.chi2_contingency(observed)
```

```python
#output values
print('Observed')
print(observed.values)
print('\nExpected')
print(expected.astype(int))
print('\n----')
print(f'chi^2 = {chi2:.4f}')
print(f'p     = {p:.4f}')
```

```python
#compare our p-value and alpha
if p < alpha:
    print('reject the null hypothesis')
else:
    print('fail to reject the null hypothesis')
```

```python
# we can conclude there is a relationship between macbook users and code up students.
```

A sample of 50 sales from office #2 revealed a mean of 100 days and a standard deviation of 20 days. Use a .05 level of significance.
